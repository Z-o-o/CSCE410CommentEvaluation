```python
from google.cloud import bigquery
from bq_helper import BigQueryHelper
import pandas as pd
```

```python
stackOverflow = BigQueryHelper(active_project="bigquery-public-data", dataset_name="stackoverflow")
bq_assistant = BigQueryHelper("bigquery-public-data", "stackoverflow")
```

```python
query = """SELECT q.id AS question_id, q.answer_count AS question_answer_count, q.accepted_answer_id AS question_accepted_answer_id, q.tags AS question_tags, q.title AS question_title, q.body AS question_body, a.id AS answer_id, a.body AS answer_body, a.parent_id AS answer_parent_id FROM bigquery-public-data.stackoverflow.posts_questions AS q JOIN bigquery-public-data.stackoverflow.posts_answers a ON q.id=a.parent_id WHERE q.tags LIKE '%python'"""
bq_assistant.estimate_query_size(query)
```

```python
response = stackOverflow.query_to_pandas_safe(query, max_gb_scanned=70)
```

```python
import sys
sys.getsizeof(response)
```

```python
response.tail(10)
```

```python
response.to_pickle("python_database.pkl")
```
